```python
breakfast= pd.read_csv('breakfast.csv')
breakfast
```

```python
%%time
from urllib import request
import io
from PIL import Image
identifier = 0

for id in breakfast.index:
    url = breakfast['Фото'].values[id]
    url = url.replace('"','')
    url = url.replace('[','')
    url = url.replace(']','')
    url = url.replace(',','')

    urls = url.split(' ')
    try:
        for item in urls:
            identifier += 1
            with request.urlopen(item) as link:
                f = io.BytesIO(link.read())
            img = Image.open(f)
            img.save('breakfast2/'+str(identifier)+'.jpg')
    except:
        continue
```
